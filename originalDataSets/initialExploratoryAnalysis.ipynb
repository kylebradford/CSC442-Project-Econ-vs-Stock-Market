```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(r'C:\Users\kjbsh\Documents\CSC442-Project-Econ-vs-Stock-Market\finalMerged', parse_dates=['Date'])

# Standardize column names
df.columns = df.columns.str.strip()  # Remove extra spaces

# Filter only Germany
df_germany = df[df['Country'] == 'Germany']

# Ensure data is sorted by date
df_germany = df_germany.sort_values(by='Date')

# Plot GDP and Close values
plt.figure(figsize=(10, 5))
plt.plot(df_germany['Date'], df_germany['<CLOSE>'], label='Close', color='tab:blue')
plt.plot(df_germany['Date'], df_germany['Gross Domestic Product (GDP)'], label='GDP', color='tab:orange')

# Labels and title
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('GDP vs. Close Over Time (Germany)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```
